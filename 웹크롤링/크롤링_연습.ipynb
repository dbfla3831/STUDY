# 라이브러리 불러오기

```python
import urllib.request # url을 가져오기 위해
import json # 입력 받을 때 json으로 받기 위해
import pandas as pd

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

# 네이버 검색 API 연습

```python
# 자신의 api 등록 정보
client_id = 'client_id' 
client_secret = 'client_secret'

# 검색할 단어, 입력 수, 페이지수, 정렬
word = input('검색할 뉴스 내용을 입력하세요 : ')
num = int(input('한 번에 표시할 검색 결과수를 입력하세요 : '))
page = int(input('원하는 페이지 번호를 입력하세요 : '))
sort = input('정렬 순서를 입력하세요 (정확도 : 1, 최신순 : 2) : ')

# 클라이언트가 서버에 요청할 url 정보 생성
enctext = urllib.parse.quote(word) # 단어를 utf로 변경
# print(enctext)

# 정렬 순서를 값으로 변경(1 : sim, 2 : date)
if sort == '1':
    sort = 'sim'
else:
    sort = 'date'
    
# 뉴스는 news, 쇼핑은 shop, 이미지는 image,,,,
url = 'https://openapi.naver.com/v1/search/news.json?query={}&display={}&start={}&sort={}'.format(enctext, num, page, sort) # GET 방식이므로 '?뒤에 변수 = 값 형식'으로 지정
# print(url)

# URL 정보로 서버에게 요청과 응답
request = urllib.request.Request(url) # 요청
request.add_header('X-Naver-Client-Id', client_id) # 자신의 api정보를 header에 저장
request.add_header('X-Naver-Client-Secret', client_secret)

response = urllib.request.urlopen(request) # 응답

# 응답을 잘 받았는지 확인
rescode = response.getcode()

if rescode == 200:
    response_body = response.read() # read : 응답의 바디 읽어오기
    print(response_body.decode('utf-8')) # 응답을 utf-8로 변경하여 출력
else:
    print('error code : ', rescode) # 에러 발생 시 에러 코드 출력
    
# json 분석
json_object = json.loads(response_body) # json객체로
```

```python
# title만 가져오기 -> items 안에 title

for title in json_object['items']:
    print(title['title'])
```

```python
# 제목만 DataFrame에 저장하기

df = pd.DataFrame() # 빈 DataFrame 생성
lst_title = [] # title을 넣을 빈 리스트 생성

# lst_title에 title 넣기
for title in json_object['items']:
    lst_title.append(title['title'])

# df에 데이터 저장
df['title'] = lst_title
df

# DataFrame을 cvs로 저장
df.to_csv('title.csv', index = False, encoding = 'utf-8')
```

```python
# DataFrame에 제목, 내용, link데이터 저장하기
df = pd.DataFrame(json_object['items'])
df = df[['title', 'description', 'link']]

# 필요없는 문자열 지우기
def str_del(df):
    for col in df.columns:
        df[col] = df[col].str.replace('</b>', '')
        df[col] = df[col].str.replace('<b>', '')   
        df[col] = df[col].str.replace('\xa0', '')
    return df
    
df = str_del(df)
df

# DataFrame을 cvs로 저장
df.to_csv('news.csv', index = False, encoding = 'utf-8')
```
