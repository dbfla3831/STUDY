```python
import pandas as pd
import os
```

```python
def read_excel_files(directory):
    dfs = []
    # 디렉토리 내의 모든 파일과 디렉토리를 반복
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.xlsx'):
                file_path = os.path.join(root, file)
                # Excel 파일을 데이터프레임으로 읽기
                df = pd.read_excel(file_path)
                # 필요한 열만 선택하여 데이터프레임에 추가
                selected_cols = df[['Unnamed: 3', 'Unnamed: 1']]
                dfs.append(selected_cols)
    # 모든 데이터프레임을 하나로 합치기
    combined_df = pd.concat(dfs, ignore_index=True)
    return combined_df

# Excel 파일이 있는 디렉토리 경로
directory = r'C:\Users\mit005\Desktop\python\pro\선거예측\data\2024people'

# 모든 Excel 파일을 읽어서 선택한 열만 포함하는 데이터프레임으로 합치기
combined_df = read_excel_files(directory)

# 결과 출력
combined_df
```

```python
combined_df = combined_df.dropna(axis = 0)
```

```python
combined_df = combined_df[~combined_df['Unnamed: 1'].str.contains('소속')]
combined_df.columns = [['name', 'party']]
```
