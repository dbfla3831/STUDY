##  20L가격과 5L가격이 모두 0원이 아닌 데이터만 필터를 한 후, 각 row별로 20L가격과 5L가격의 차이를 ‘차이가격’ 이라 부른다고 하자. 시도명 별 차이가격의 평균가격을 비교할때 그 값이 가장 큰 금액을 반올림하여 소숫점 이하 1자리까지 구하여라

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_1_.csv')
df.head(5)
```

```python
df = df[(df['20L가격'] != 0) & (df['5L가격'] != 0)]
res_df = df.copy()
res_df['차이가격'] = res_df['20L가격'] - res_df['5L가격']
res_gb = res_df.groupby('시도명')['차이가격'].mean().sort_values(ascending = False)[0]
res = round(res_gb, 1)
print(res)
```

##  BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 초고도 비만은 BMI 25이상 , 고도 비반은 BMI 25미만 - 23이상 , 정상은 23미만 - 18.5이상 저체중은 18.5미만으로 정의 된다. 주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
df.head(5)
```

```python
df['bmi'] = df['weight(kg)'] / (df['height(cm)'] / 100) ** 2

def lavel(x):
    if x >= 25:
        return '초고도비만'
    elif 23 <= x < 25:
        return '고도비만'
    elif 18.5 <= x < 23:
        return '정상'
    else:
        return '저체중'
df['BMI'] = df['bmi'].apply(lavel)
res = len(df[(df['BMI'] == '초고도비만') | (df['BMI'] == '저체중')])
print(res)
```

##  순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내, 도외 전출인원을 뺀값이다. 각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 전체 기간의 해당 순유입인원들의 합을 구하여라

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_3.csv')
df.head(5)
```

```python
df['순유입인원'] = df.iloc[:, [3, 4, 7, 8, 11, 12]].sum(axis = 1) - df.iloc[:, [1, 2, 5, 6, 9, 10]].sum(axis = 1)
res_gb = df.groupby(['년도', '지역'])['순유입인원'].sum().reset_index().sort_values(['년도', '순유입인원'], ascending = False).groupby('년도').head(1)
res = res_gb['순유입인원'].sum()
print(res)
```

# 작업 2유형
예측 변수 price, test.csv에 대해 ID별로 price 값을 예측하여 제출, 제출 데이터 컬럼은 ID와 price 두개만 존재해야함. 평가지표는 rmse

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

display(train.head(2))
test.head(2)
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

col_drop = ['ID']
col_y = ['price']
col_num = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
col_cat = ['fuelType', 'model', 'transmission']
test_id = test['ID']

train = train.drop(col_drop, axis = 1)
test = test.drop(col_drop, axis = 1)

x_train = train.loc[:, ['model', 'year', 'transmission', 'mileage',	'fuelType',	'tax', 'mpg', 'engineSize']]
y_train = train.loc[:, ['price']]
x_test = test

x = pd.concat([x_train, x_test])
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown = 'ignore')
ohe.fit(x[col_cat])

x_train_res = ohe.transform(x_train[col_cat])
x_test_res = ohe.transform(x_test[col_cat])

x_train_ohe = pd.DataFrame(x_train_res.todense(), columns = ohe.get_feature_names_out())
x_test_ohe = pd.DataFrame(x_test_res.todense(), columns = ohe.get_feature_names_out())

x_train_fin = pd.concat([x_train_ohe, x_train[col_num]], axis = 1)
x_test_fin = pd.concat([x_test_ohe, x_test[col_num]], axis = 1)

from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(x_train_fin, y_train.values.ravel(), test_size = 0.3)#, stratify = y_train.values.ravel())

from sklearn.preprocessing import MinMaxScaler
x_tr = X_tr.copy()
x_val = X_val.copy()
X_test_fin = x_test_fin.copy()

scaler = MinMaxScaler()
scaler.fit(x_tr[col_num])

x_tr[col_num] = scaler.transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
X_test_fin[col_num] = scaler.transform(X_test_fin[col_num])

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_tr, y_tr)


from sklearn.metrics import mean_squared_error
y_tr_pred = model.predict(x_tr)
y_val_pred = model.predict(x_val)

print('tr rmse : ', mean_squared_error(y_tr, y_tr_pred, squared = False))
print('val rmse : ', mean_squared_error(y_val, y_val_pred, squared = False))

y_pred = model.predict(X_test_fin)
result = pd.DataFrame({'ID' : test_id, 'price' : y_pred})
result
```

# 작업 3유형

## 어느 학교에서 추출한 55명 학생들의 키 정보이다. 이 학생들의 키의 95% 신뢰구간을 구하고자 한다

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_1.csv')
df.head()
```

###  55명 학생들의 키에 대한 표본 평균을 구하여라(반올림하여 소숫점 3째자리까지)

```python
res = round(df['height'].mean(), 3)
print(res)
```

###  t분포 양쪽 꼬리에서의 t값을 구하여러(반올림하여 소수 4째자리)

```python
from scipy.stats import t

t = t.ppf((1 + 0.95)/2, 55-1)
print(round(t, 4))
```

t.ppf(유의수준, 자유도)

### 95% 신뢰구간을 구하여라(print(lower, upper)방식으로 출력, 각각의 값은 소숫점 이해 3째자리까지)

```python
import numpy as np
upper = df['height'].mean() + (t * df['height'].std() / np.sqrt(55))
lower = df['height'].mean() - (t * df['height'].std() / np.sqrt(55))
print(round(lower, 3), round(upper, 3))
```

## A,B,C 세 공장에서 생산한 동일한 제품의 길이 데이터이다. 공장간의 제품 길이 차이가 유의미한지 확인 하려한다

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_2.csv')
df.head()
```

### 3 그룹의 데이터에 대해 크루스칼-왈리스 검정을 사용하여 검정 통계량을 반올림하여 소숫점 이하 3자리까지 구하여라

```python
from scipy.stats import kruskal
a = df[df['ID'] == 'A']['value']
b = df[df['ID'] == 'B']['value']
c = df[df['ID'] == 'C']['value']

t, p = kruskal(a, b, c)
res = round(t, 3)
print(res)
```

### . 3 그룹의 데이터에 대해 크루스칼-왈리스 검정을 사용하여 p-value를 반올림하여 소숫점 이하 3자리까지 구하여라. 귀무가설과 대립가설중 0.05 유의수준에서 유의한 가설을 출력하라

```python
from scipy.stats import kruskal
a = df[df['ID'] == 'A']['value']
b = df[df['ID'] == 'B']['value']
c = df[df['ID'] == 'C']['value']

t, p = kruskal(a, b, c)
res = round(p, 3)
print(res)

if res < 0.05:
    print('기각')
else:
    print('채택')
```
