# 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
df.head()
```

## 성별이 Male인 환자들의 age의 평균값은 ?

```python
df['age'] = df['age'].str.replace('*', '').astype('int')
res = df[df['gender'] == 'Male']['age'].mean()
print(res)
```

## bmi컬럼의 결측치를 bmi컬럼의 결측치를 제외한 나머지 값들의 중앙값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

```python
med = df[df['bmi'].notnull()]['bmi'].median()
res_df = df.copy()
res_df['bmi'] = res_df['bmi'].fillna(med)
res = round(res_df['bmi'].mean(), 3)
print(res)
```

## bmi컬럼의 각 결측치들을 직전의 행의 bmi값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

```python
res_df = df.copy()
res_df['bmi'] = res_df['bmi'].fillna(method = 'ffill')
res = round(res_df['bmi'].mean(), 3)
print(res)
```

## bmi컬럼의 각 결측치들을 결측치를 가진 환자 나이대(10단위)의 평균 bmi 값으로 대체한 후 대체된 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

```python
# res_df = df.copy()
# res_df['age'] = res_df['age'] // 10 * 10

# nan_df = res_df[res_df['bmi'].notnull()]
# avg_gb = nan_df.groupby('age')['bmi'].mean()

# dic = { x:y for x,y in avg_gb.items()}

# res_df.loc[res_df['bmi'].isnull()]['bmi'] = res_df[res_df['bmi'].isnull()]['age'].map(lambda x: dic[x])
# res = res_df['bmi'].mean()
# print(res)

mean = df[df.bmi.notnull()].groupby(df.age//10 *10).bmi.mean()
dic = { x:y for x,y in mean.items()}
df.loc[df.bmi.isnull(),'bmi']  =(df[df.bmi.isnull()].age//10*10).map(lambda x : dic[x])


result = df.bmi.mean()
print(result)
```

## avg_glucose_level 컬럼의 값이 200이상인 데이터를 모두 199로 변경하고 stroke값이 1인 데이터의 avg_glucose_level값의 평균을 소수점이하 3자리 까지 구하여라

```python
df.loc[df.avg_glucose_level >=200,'avg_glucose_level'] =199
res = round(df[df['stroke'] == 1]['avg_glucose_level'].mean(), 3)
print(res)
```

# 작업 1유형_다른 데이터

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/pok/Pokemon.csv')
df.head()
```

## Attack컬럼의 값을 기준으로 내림차순정렬 했을때 상위 400위까지 포켓몬들과 401~800위까지의 포켓몬들에서 전설포켓몬(Legendary컬럼)의 숫자 차이는?

```python
res_df = df.copy()
res_df = res_df.sort_values('Attack', ascending = False).reset_index(drop = True)
df_400 = res_df.loc[:400]
df_401 = res_df.loc[401:]

l_400 = len(df_400[df_400['Legendary'] == True])
l_401 = len(df_401[df_401['Legendary'] == True])

res = l_400 - l_401
print(res)
```

## Type 1 컬럼의 종류에 따른 Total 컬럼의 평균값을 내림차순 정렬했을때 상위 3번째 Type 1은 무엇인가?

```python
res_gb = df.groupby('Type 1')['Total'].mean().sort_values(ascending = False)
res = res_gb.index[2]
print(res)
```

## 결측치가 존재하는 행을 모두 지운 후 처음부터 순서대로 60% 데이터를 추출하여 Defense컬럼의 1분위수를 구하여라

```python
res_df = df.copy()
res_df = res_df.dropna().reset_index(drop = True)
res_df = res_df.loc[:len(res_df) * 0.6]
res = res_df['Defense'].quantile(0.25)
print(res)
```

## type 1 컬럼의 속성이 Fire인 포켓몬들의 Attack의 평균이상인 Water속성의 포켓몬 수를 구하여라

```python
avg = df[df['Type 1'] == 'Fire']['Attack'].mean()
res_df = df[(df['Type 1'] == 'Water') & (df['Attack'] >= avg)]
res = len(res_df)
print(res)
```

## 각 세대 중(Generation 컬럼)의 Speed와 Defense 컬럼의 차이(절댓값)이 가장 큰 세대는?

```python
res_gb = df.groupby('Generation')[['Speed', 'Defense']].sum()
res_gb['diff'] = abs(res_gb['Speed'] - res_gb['Defense'])
res_gb = res_gb.sort_values('diff', ascending = False)
res = res_gb.index[0]
print(res)
```

## 방법 1. dummies, RandomForestClassifier

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/test.csv')


train['bmi'] = train['bmi'].fillna(train['bmi'].mean())
test['bmi'] = test['bmi'].fillna(test['bmi'].mean())
train['age'] = train['age'].str.replace('*', '').astype('int')
# test['age'] = test['age'].str.replace('*', '').astype('int')


col_drop = ['id']
col_num = ['hypertension', 'heart_disease', 'avg_glucose_level', 'bmi', 'age']
col_cat = ['Residence_type', 'ever_married', 'gender', 'smoking_status', 'work_type']
col_y = 'stroke'
test_id = test['id']

train = train.drop(col_drop, axis = 1)
test = test.drop(col_drop, axis = 1)

x_train = train.loc[:, 'gender' : 'smoking_status']
x_test = test
y_train = train[col_y]

x_train_dum = pd.get_dummies(x_train)
x_test_dum = pd.get_dummies(x_test)

diff_train = set(x_train_dum.columns) - set(x_test_dum.columns)
for col in diff_train:
    x_test_dum[col] = 0

diff_test = set(x_test_dum.columns) - set(x_train_dum.columns)
for col in diff_test:
    x_train_dum[col] = 0

x_test_dum = x_test_dum[x_train_dum.columns]

from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(x_train_dum, y_train.values.ravel(), test_size = 0.3)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_tr[col_num])

x_tr[col_num] = scaler.transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
x_test_dum[col_num] = scaler.transform(x_test_dum[col_num])

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_tr, y_tr)

from sklearn.metrics import accuracy_score, roc_auc_score
y_tr_pred = model.predict(x_tr)
y_val_pred = model.predict(x_val)
print('ac tr : ', accuracy_score(y_tr_pred, y_tr))
print('ac val : ', accuracy_score(y_val_pred, y_val))

y_tr_pred = model.predict_proba(x_tr)[:, 1]
y_val_pred = model.predict_proba(x_val)[:, 1]
print('rac tr : ', roc_auc_score(y_tr, y_tr_pred))
print('rac val : ', roc_auc_score(y_val, y_val_pred))

y_pred = model.predict_proba(x_test_dum)[:, 1]
result = pd.DataFrame({'id' : test_id, 'y_pred' : y_pred})
result.to_csv('0030.csv', index = False)
```
