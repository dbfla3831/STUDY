# 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
df.head()
```

## 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는? (0~9 : 0 , 10~19 : 10)

```python
df['age_label'] = df['age'] // 10 * 10
res_gb = df.groupby('age_label').size().sort_values(ascending = False)
res = res_gb.index[0]
print(res)
```

## 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 나이대 구간의 인원은 몇명인가?

```python
res_gb = df.groupby('age_label').size().sort_values(ascending = False)
res = res_gb.values[0]
print(res)
```

## 나이가 25살 이상 29살 미만인 응답 고객들중 housing컬럼의 값이 yes인 고객의 수는?

```python
res_df = df[(df['age'] >= 25) & (df['age'] < 29)]
res_df = res_df[res_df['housing'] == 'yes']
res = len(res_df)
print(res)
```

## numeric한 값을 가지지 않은 컬럼들중 unique한 값을 가장 많이 가지는 컬럼은?

```python
res_df = df.select_dtypes(include = 'object')
res_gb = res_df.nunique().sort_values(ascending = False)
res = res_gb.index[0:2]
print(res)
```

## balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID값을 기준으로 내림차순 정렬했을때 상위 100개 데이터의 balance값의 평균은?

```python
res_df = df[df['balance'] >= df['balance'].mean()]
res_df = res_df.sort_values('ID', ascending = False).iloc[:100]
res = res_df['balance'].mean()
print(res)
```

## 가장 많은 광고를 집행했던 날짜는 언제인가? (데이터 그대로 일(숫자),달(영문)으로 표기)

```python
res_gb = df.groupby(['day', 'month']).size().sort_values(ascending = False)
res = res_gb.index[0]
print(res)
```

## 데이터의 job이 unknown 상태인 고객들의 age 컬럼 값의 정규성을 검정하고자 한다. 샤피로 검정의 p-value값을 구하여라

```python
res_df = df[df['job'] == 'unknown']['age']

from scipy.stats import shapiro
t_val, p_val = shapiro(res_df)

print(p_val)
```

## age와 balance의 상관계수를 구하여라

```python
corr_df = df[['age', 'balance']].corr()
res = corr_df.iloc[0, 1]
print(res)
```

## y 변수와 education 변수는 독립인지 카이제곱검정을 통해 확인하려한다. p-value값을 출력하라

```python
from scipy.stats import chi2_contingency

data = pd.crosstab(df['y'], df['education'])
t, p, _, _ = chi2_contingency(data)
print(p)
```

## 각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?

```python
div = df[df['marital'] == 'divorced'].groupby('job').size().to_frame().reset_index()
ma = df[df['marital'] == 'married'].groupby('job').size().to_frame().reset_index()
res_gb = pd.merge(div, ma, on = 'job')
res_gb['ratio'] = res_gb['0_x'] / res_gb['0_y']
res_gb = res_gb.sort_values('ratio', ascending = False).reset_index(drop = True)
res = res_gb.loc[0, 'ratio']
print(res)
```
