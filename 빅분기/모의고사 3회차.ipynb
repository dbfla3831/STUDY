# 작업 1유형

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/spotify/spotify.csv')
df.head()
```

## 데이터는 현재 년도별 100곡이 인기순으로 정렬되어 있다. 각 년도별 1~100위의 랭킹을 나타내는 rank컬럼을 만들고 매년도 1위의 bpm컬럼의 평균값을 구하여라

```python
df = df.dropna()
df.loc[:, 'rank'] = list(range(1, 101, 1)) * 10
res = df[df['rank'] == 1]['bpm'].mean()
print(res)
```

## 2015년도에 가장많은 top100곡을 올린 artist는 누구인가?

```python
res_df = df[df['top year'] == 2015]
res = res_df.groupby('artist').size().sort_values(ascending = False).index[0:2]
print(res)
```

## 년도별 rank값이 1~10위 까지의 곡들 중 두번째로 많은 top genre는 무엇인가?

```python
df.head(2)
```

```python
res_df = df[(df['rank'] >= 1) & (df['rank'] <= 10)]
res_gb = res_df.groupby('top genre').size().sort_values(ascending = False).index[1]
print(res_gb)
```

## 피처링의 경우 title에 표시된다. 피처링을 가장 많이 해준 가수는 누구인가?

```python
df['feat'] = df['title'].str.split('feat. ').str[1]
df['feat'] = df['feat'].str.replace(')', '')
res_df = df.dropna()
res = res_df.groupby('feat').size().sort_values(ascending = False).index[0]
print(res)
```

## top year 년도를 기준으로 발매일(year released)과 top100에 진입한 일자 (top year)가 다른 곡의 숫자를 count 했을때 가장 많은 년도는?

```python
res_df = df[df['top year'] != df['year released']]
res = res_df.groupby('top year').size().sort_values(ascending = False).index[0]
print(res)
```

## artist 컬럼의 값에 대소문자 상관없이 q 단어가 들어가는 아티스트는 몇명인가?

```python
df['artist'] = df['artist'].str.lower()
res_df = df[df['artist'].str.contains('q')]
res = res_df.groupby('artist').size().sort_values(ascending = False)
print(len(res))
```

## 년도 상관없이 전체데이터에서 1 ~ 50위와 51 ~ 100위간의 dur 컬럼의 평균값의 차이는?

```python
m1 = df[df['rank'].isin(range(1, 51))]['dur'].mean()
m2 = df[df['rank'].isin(range(51, 101))]['dur'].mean()

res = m1 - m2
print(res)
```

## title을 띄어쓰기 단어로 구분 했을때 가장 많이 나온 단어는 무엇인가? (대소문자 구분 x)

```python
res_df = df.copy()
res_df['title'] = res_df['title'].str.split(' \(feat').str[0]
res = res_df['title'].str.split().explode().str.lower().value_counts().sort_values(ascending = False).index[0]
res
```

## 년도별 nrgy값의 평균값을 구할때 최대 평균값과 최소 평균값의 차이를 구하여라

```python
res_df = df.groupby('top year')['nrgy'].mean().sort_values(ascending = False)
res = res_df.values[0] - res_df.values[-1]
print(res)
```

## artist중 artist type 타입을 여러개 가지고 있는 artist는 누구인가

```python
res_df = df[['artist', 'artist type']].value_counts().reset_index().artist.value_counts().sort_values(ascending = False).index[0]
print(res_df)
```

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/y_train.csv")
test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_test.csv")

# null, 이상치 X, 명목형 X

col_drop = ['ID']
test_id = test['ID']

x_train = x_train.drop(col_drop, axis = 1)
y_train = y_train.drop(col_drop, axis = 1)
x_test = test.drop(col_drop, axis = 1)

from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train.values.ravel(), test_size = 0.3, stratify = y_train.values.ravel())

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_tr)

x_tr = scaler.transform(x_tr)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_tr, y_tr)

from sklearn.metrics import accuracy_score, roc_auc_score
y_tr_pred = model.predict(x_tr)
y_val_pred = model.predict(x_val)

print('as tr : ', accuracy_score(y_tr_pred, y_tr))
print('as val : ', accuracy_score(y_val_pred, y_val))

y_tr_pred = model.predict_proba(x_tr)[:, 1]
y_val_pred = model.predict_proba(x_val)[:, 1]

print('ras tr : ', roc_auc_score(y_tr, y_tr_pred))
print('ras val : ', roc_auc_score(y_val, y_val_pred))

y_pred = model.predict_proba(x_test)[:, 1]
result = pd.DataFrame({"ID" : test_id, 'pose' : y_pred})
result.to_csv('00300.csv', index = False)
```
