# 라이브러리 불러오기

```python
import pandas as pd
```

# 데이터 불러오기

```python
df = pd.read_csv('./data/ex1.csv')
df
```

pd.read_csv의 파라미터

- 구분자

        sep = '구분문자'

- 인코딩 

         encoding = 'utf-8', 'cp949'
         
- header
        
        header = column이름으로 지정할 행 지정
                 첫 행이 헤더가 아닌 경우 header = None
         
- 특정 columns를 index로 지정

        index_col = 'column명' / 인덱스 번호
        
- column명 변경

        namea = []
        


```python
# col이 없는 경우

pd.read_csv('./data/ex2.csv')
```

```python
# col은 none으로

pd.read_csv('./data/ex2.csv', header = None)
```

```python
# col에 이름 지정

pd.read_csv('./data/ex2.csv', names = ['a', 'b', 'c', 'd', 'message'])
```

```python
# index, col에 이름 지정

names = ['a', 'b', 'c', 'd', 'message']
pd.read_csv('./data/ex2.csv', names = names, index_col = 'message')
```

```python
# 멇티 index

pd.read_csv('./data/csv_mindex.csv', index_col = ['key1', 'key2'])
```

```python
# 공백 문자로 분할

pd.read_csv('./data/ex3.txt', sep = '\s+')
```

```python
# 원하는 만큼 불러오기

pd.read_csv('./data/ex6.csv', nrows = 5)
```

# 데이터 저장

```python
data = pd.read_csv('./data/ex5.csv')
data

data.to_csv('out.csv') # csv로 저장 .to_csv('파일명.확장자')
```

<위 실행 결과>
,something,a,b,c,d,message

0,one,1,2,3.0,4,

1,two,5,6,,8,world

2,three,9,10,11.0,12,foo


```python
# 저장할 때 누락된 값을 원하는 값으로 변경

data.to_csv('out.csv', na_rep = '없음')
```

<위 실행 결과>

,something,a,b,c,d,message

0,one,1,2,3.0,4,없음

1,two,5,6,없음,8,world
