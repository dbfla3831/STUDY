## 5.집합


- 집합은 순서가 없고 중복이 없다

```python
a = {1, 2, 3, 2, 1}
b = set([1, 2, 3, 2, 1])
c = set('python')

print(a)
print(b)
print(c)
```

```python
a = set([1, 2, 3])
b = list(a)
c = tuple(a)

print(a)
print(b)
print(b[0])
print(c)
print(c[2])
```

```python
# 교집합
a = set([1, 2, 3, 4, 5])
b = set([3, 4, 5, 6, 7])
c = a & b
d = a.intersection(b)

print(c)
print(d)


# 합집합
a = set([1, 2, 3, 4, 5])
b = set([3, 4, 5, 6, 7])
c = a | b
d = a.union(d)

print(c)
print(d)


# 차집합
a = set([1, 2, 3, 4, 5])
b = set([3, 4, 5, 6, 7])
c = a - b
d = a.difference(b)

print(c)
print(d)
```

## 3.scikit - learn 머신러닝 만들어 보기

### 1.데이터 불러오기

```python
import pandas as pd

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['label'] = iris.target
df.head()
```

### 2.데이터 나누기

```python
from sklearn.model_selection import train_test_split

data = iris.data
target = iris.target

x_train, x_test, y_train, y_test = train_test_split(data, target, test_size = 0.2, random_state = 11)

print('train_dataset')
print('x_train : ', len(x_train))
print('y_train : ', len(y_train))

print()

print('test_dataset')
print('x_tset : ', len(x_test))
print('y_test : ', len(y_test))
```

### 3.모델 학습

```python
from sklearn.tree import DecisionTreeClassifier

df_clf = DecisionTreeClassifier(random_state = 11)
df_clf.fit(x_train, y_train)
```

### 4.모델 예측

```python
from sklearn.metrics import accuracy_score

pred =df_clf.predict(x_test)

ac_score = accuracy_score(y_test, pred)

print("예측 정확도 : ", ac_score)
```

### 1.레이블 인코딩

- Label Encoder를 생성 한 후 fit와 transform을 호출 후 레이블 인코딩을 수행

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', '컴퓨터', 'TV','냉장고', '컴퓨터', '컴퓨터']

encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)

print(labels)
print(encoder.classes_)
```

```python
origns = encoder.inverse_transform(labels)

print(origns)
```

### 2.원 - 핫 인코딩

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

labels = labels.reshape(-1, 1)

oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)
print(oh_labels.toarray())
```

### 3.스케일링

- 표준화 : (x - 해당 데이터 세트의 평균) / 데이터 세트의 표준편자

- 정규화 : 최대 - 최소 정규화 (x - 최소) / (최대 - 최소)

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
data = iris.data
df = pd.DataFrame(data, columns = iris.feature_names)

print(df.mean())
print()
print(df.var())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df)

iris_scaler = scaler.transform(df)
iris_scaler_df = pd.DataFrame(iris_scaler, columns = iris.feature_names)

print(iris_scaler_df.mean())
print(iris_scaler_df.var())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df)

iris_scaler = scaler.transform(df)
iris_scaler_df = pd.DataFrame(iris_scaler, columns = iris.feature_names)

print(iris_scaler_df.min())
print(iris_scaler_df.max())
```

- 학습 데이터로 fit이 적용된 스케일링 기준 정보를 그대로 테스트 데이터에 적용해야 한다

-> 테스트 데이터에 다시 fit을 적용해서는 안 된다

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

train_array = np.arange(0, 10).reshape(-1, 1)
test_array = np.arange(0, 6).reshape(-1, 1)

scaler = MinMaxScaler()
scaler.fit(train_array)
train_scaler = scaler.transform(train_array)

print(np.round(train_array.reshape(-1), 2))
print(np.round(train_scaler.reshape(-1), 2))

print()

# scaler.fit(test_array)
test_scaler = scaler.transform(test_array)

print(np.round(test_array.reshape(-1), 1))
print(np.round(test_scaler.reshape(-1), 1))
```
